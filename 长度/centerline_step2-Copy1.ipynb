```python
def glength(skelpath,dirin,txt ):
    ext = np.round(extofraster(skelpath)).astype(int)
    x = (ext[0]+ext[2])/2
    y = (ext[1]+ext[3])/2
    grwlvec = grwlext(ext,dirin,txt, extt = '.shp')
    riverlength = 0
    for grwlpath in grwlvec:
        l1 = grwllen(grwlpath, ext)
        riverlength = riverlength + l1
    return riverlength, x,y 
```

```python
def grwllen(grwlpath, ext):
    river = gpd.read_file(grwlpath)
    river = river[river['lakeFlag'] != 1]
    river = river[(river['lon'] > ext[0]) & (river['lon'] < ext[2]) & (river['lat'] > ext[1]) & (river['lat'] < ext[3])]

    if len(river) != 0:
        length = cal_length(river)['length'].sum()
    else: 
        length = 0
    return length
```

```python
import numpy as np
from scipy.ndimage import convolve

def count_ones_around(arr):
    # 定义3x3的卷积核
    kernel = np.array([[1, 1, 1],
                       [1, 0, 1],
                       [1, 1, 1]])

    # 使用卷积操作计算每个位置周围为1的总和
    result = convolve(arr, kernel, mode='constant', cval=0)

    return result

import os
from chu import raster2shp1
def change_suffix(file_path, new_suffix):
    # 分离文件名和后缀
    base_path, old_suffix = os.path.splitext(file_path)
    
    # 构建新的文件路径
    new_file_path = base_path + new_suffix
    
    return new_file_path

import numpy as np
import os
def rasterwrite(temp, data, pro):
    try:
        with rasterio.open(temp,'w',**pro) as dst:
            dst.write(data)
    except ValueError:
        data = data[np.newaxis,:]
        with rasterio.open(temp,'w',**pro) as dst:
            dst.write(data)
            
def array2shp(data, temp ,pro):
    #temp = os.path.join(dirtemp, prefix+os.path.split(intif)[1])
    #temp = os.path.join(dirtemp, 'jdis_'+os.path.split(intif)[1])
    rasterwrite(temp,data,pro)
    raster2shp1(temp,1,'')
    shppath = change_suffix(temp, '.shp')
    os.remove(temp)
    return shppath

def je_dis(skelpath, dispath, dirtemp):
    with rasterio.open(skelpath) as src1, rasterio.open(dispath) as src2:
        skel = src1.read(1)
        pro1 = src1.profile
        
        dis = src2.read(1)
        pro2 = src2.profile
    if pro1 == pro2:
        pro = pro1
    else:
        print('error: pro1 != pro2')
        pro = 0
    nearby_points = count_ones_around(skel)
    nbp = nearby_points*skel
    joints = np.where(nbp >2 , 1, 0)
    ends = np.where(nbp ==1 , 1, 0)
    j_dis = joints*dis
    e_dis = ends*dis
    
    tempj = os.path.join(dirtemp, 'jdis1_'+os.path.split(skelpath)[1])
    tempe = os.path.join(dirtemp, 'edis1_'+os.path.split(skelpath)[1])
    shpj = array2shp(j_dis, tempj, pro)
    shpe = array2shp(e_dis, tempe, pro)
    return shpj, shpe, pro
```

```python
def debranchlen(linespath, jdis, edis, pa1 = 1, pa2 = 1):
    #前提是lines不能为空， linespath 是skel 在gis里面转线后的shp；  lines为空即skel全部为0
    lines = gpd.read_file(linespath)
    lines_p = cal_length(lines)
    if pa1 == 1:
        jdis = gpd.read_file(jdis)
    if pa2 == 1:
        edis = gpd.read_file(edis)
    if len(jdis) !=0:
        
        linejj = class1(lines_p,jdis)
        lines_pj = class2(lines_p, jdis,edis)
        lines_without_intersections = debranch(lines_p, linejj,lines_pj)
        #length = lines_without_intersections['length'].sum()
    else:
        lines_without_intersections = lines_p
    
    length = lines_without_intersections['length'].sum()
    return lines_without_intersections, length
```

```python
def cal_length(lines):
    lines_p = lines.to_crs(epsg=3857) 
    lines_p['length'] = lines_p['geometry'].length
    lines_p = lines_p.to_crs(epsg=4326) 
    return lines_p

#（j-j中小于15m的，j-e中小于j长度的1.42倍的； 考虑：j-e中小于300m的，太短的分支不要；保留全部e-e的）
def class1(lines_p,jdis):
    step1 = lines_p[lines_p['length'] <15]
    step2 = step1.sjoin(jdis, how='inner', predicate='intersects')
    #找到长度小于15m的，并确定这个line 是两个joints作为端点  
    return step2

def class2(lines_p, jdis,edis, num = 300):
    lines_pj = lines_p.sjoin(jdis, how='inner', predicate='intersects')  #(j-j和j-e)
    lines_pj['DN'] = lines_pj['DN']*14.2
    lines_pj = lines_pj.rename(columns={'DN': 'dis'})
    lines_pj = lines_pj.drop(columns=['index_right'])
    
    lines_pj = lines_pj.sjoin(edis, how='inner', predicate='intersects') #(j-e)
    #lines_pj['DN'] = lines_pj['DN']*14.2
    #lines_pj = lines_pj.rename(columns={'DN': newname})
    lines_pj = lines_pj.drop(columns=['index_right'])
    # if num = 0的时候，没有河流会被长度限制而去除，否则 j-e里面长度小于num的会被删掉
    lines_pj = lines_pj[(lines_pj['length'] < lines_pj['dis'])| (lines_pj['length'] < num)]
    
    return lines_pj
```

```python
def debranch(lines_p, linejj,lines_pj):
    intersected_indices1 = set(linejj.index)
    intersected_indices2 = set(lines_pj.index)
    merged_indices = list(linejj.index.union(lines_pj.index))
    lines_without_intersections = lines_p.drop(index=merged_indices)
    return lines_without_intersections
```
